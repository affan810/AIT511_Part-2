```python
# =============== PHASE 3 : UNIFIED COMPARISON FRAMEWORK ===============
import time
import numpy as np
import pandas as pd

from sklearn.metrics import accuracy_score, f1_score

# Models
from sklearn.linear_model import LinearRegression, Ridge, Lasso, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

# Results will be appended as dicts, then converted to DataFrame
results = []
```

```python
def evaluate_model(model, name, X_test, y_test,
                   is_regression=False,
                   is_nn_proba=False):
    """
    Unified evaluation:
    - For regression models: round + clip to [1,7]
    - For neural nets (later): argmax over probabilities
    - For normal classifiers: direct predict()
    """
    start = time.time()

    if is_regression:
        y_pred_cont = model.predict(X_test)
        # In case model returns shape (n_samples, 1)
        y_pred_cont = np.ravel(y_pred_cont)
        y_pred = np.clip(np.round(y_pred_cont), 1, 7).astype(int)

    elif is_nn_proba:
        # For future NN / CNN / Transformer models
        y_proba = model.predict(X_test)
        y_pred = np.argmax(y_proba, axis=1) + 1

    else:
        y_pred = model.predict(X_test)

    end = time.time()
    elapsed = end - start

    acc = accuracy_score(y_test, y_pred)
    f1  = f1_score(y_test, y_pred, average='weighted')

    result = {
        'Model': name,
        'Accuracy': acc,
        'Weighted_F1': f1,
        'Time_Sec': elapsed
    }
    return result
```

```python
# ----------------- GROUP A: REGRESSION MODELS -----------------
# We use X_train_scaled / X_test_scaled and adapt outputs to class labels

# 1) Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X_train_scaled, y_train)
results.append(
    evaluate_model(lin_reg, "LinearRegression",
                   X_test_scaled, y_test,
                   is_regression=True)
)

# 2) Ridge Regression
ridge = Ridge(alpha=1.0)
ridge.fit(X_train_scaled, y_train)
results.append(
    evaluate_model(ridge, "Ridge",
                   X_test_scaled, y_test,
                   is_regression=True)
)

# 3) LASSO Regression
lasso = Lasso(alpha=0.001, max_iter=10000)
lasso.fit(X_train_scaled, y_train)
results.append(
    evaluate_model(lasso, "Lasso",
                   X_test_scaled, y_test,
                   is_regression=True)
)

# 4) Polynomial Regression (degree=2 as a reasonable baseline)
poly_model = Pipeline([
    ("poly", PolynomialFeatures(degree=2, include_bias=False)),
    ("linreg", LinearRegression())
])
poly_model.fit(X_train_scaled, y_train)
results.append(
    evaluate_model(poly_model, "PolynomialRegression_deg2",
                   X_test_scaled, y_test,
                   is_regression=True)
)

print("Finished Group A (Regression Models)")
```

```python
# ----------------- GROUP B: CLASSICAL CLASSIFIERS -----------------

# 1) Logistic Regression (multinomial)
log_reg = LogisticRegression(
    multi_class='multinomial',
    max_iter=1000,
    n_jobs=-1
)
log_reg.fit(X_train_scaled, y_train)
results.append(
    evaluate_model(log_reg, "LogisticRegression",
                   X_test_scaled, y_test)
)

# 2) K-Nearest Neighbors
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train_scaled, y_train)
results.append(
    evaluate_model(knn, "KNN_k7",
                   X_test_scaled, y_test)
)

# 3) Decision Tree (uses RAW data)
dt = DecisionTreeClassifier(
    max_depth=None,
    random_state=42
)
dt.fit(X_train_raw, y_train)
results.append(
    evaluate_model(dt, "DecisionTree",
                   X_test_raw, y_test)
)

# 4) Random Forest (RAW data)
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    n_jobs=-1,
    random_state=42
)
rf.fit(X_train_raw, y_train)
results.append(
    evaluate_model(rf, "RandomForest",
                   X_test_raw, y_test)
)

# 5) Gradient Boosting (RAW data)
gb = GradientBoostingClassifier(
    random_state=42
)
gb.fit(X_train_raw, y_train)
results.append(
    evaluate_model(gb, "GradientBoosting",
                   X_test_raw, y_test)
)

# 6) Support Vector Machine (RBF kernel, scaled data)
svm_rbf = SVC(
    kernel='rbf',
    C=1.0,
    gamma='scale'
)
svm_rbf.fit(X_train_scaled, y_train)
results.append(
    evaluate_model(svm_rbf, "SVM_RBF",
                   X_test_scaled, y_test)
)

print("Finished Group B (Classical Classifiers)")
```

```python
from collections import Counter

def map_clusters_to_labels(cluster_labels, y_true):
    """
    cluster_labels : array of shape (n_samples,)
    y_true         : array of shape (n_samples,) in SAME order
    """
    mapping = {}
    for c in np.unique(cluster_labels):
        # Get positions of samples in this cluster
        positions = np.where(cluster_labels == c)[0]
        labels = y_true[positions]            # <-- POSITIONS, not boolean masking
        majority = Counter(labels).most_common(1)[0][0]
        mapping[c] = majority
    return mapping
```

```python
from sklearn.mixture import GaussianMixture

results_gmm = []
```

```python
# Dictionary to store 7 trained GMM models
gmm_dict = {}

num_classes = len(np.unique(y_train))  # should be 7

for c in range(1, num_classes + 1):
    X_c = X_train_scaled[y_train == c]
    gmm = GaussianMixture(
        n_components=2,             # 2 Gaussians per class gives good balance of speed + performance
        covariance_type='full',
        max_iter=150,
        random_state=42
    )
    gmm.fit(X_c)
    gmm_dict[c] = gmm
```

```python
# Score matrix: rows = samples, columns = classes
scores = np.zeros((len(X_test_scaled), num_classes))

for i, c in enumerate(range(1, num_classes + 1)):
    scores[:, i] = gmm_dict[c].score_samples(X_test_scaled)

# Predicted class = Gaussian with highest log-likelihood
y_pred = np.argmax(scores, axis=1) + 1
```

```python
acc = accuracy_score(y_test, y_pred)
f1  = f1_score(y_test, y_pred, average='weighted')

results_gmm.append({
    "Model": "GaussianMixture_GMM_MAP",
    "Accuracy": acc,
    "Weighted_F1": f1,
    "Time_Sec": None
})

# Also add into the MAIN global results table if you prefer:
results.append({
    "Model": "GaussianMixture_GMM_MAP",
    "Accuracy": acc,
    "Weighted_F1": f1,
    "Time_Sec": None
})

print("GMM Phase Completed")
```

```python
# estimate priors from train (class frequency)
class_priors = y_train.value_counts(normalize=True).sort_index()  # class 1..7 aligned

# store 7 GMMs (1 for each class)
gmm_map_dict = {}
num_classes = len(np.unique(y_train))

for c in range(1, num_classes + 1):
    X_c = X_train_scaled[y_train == c]
    gmm = GaussianMixture(
        n_components=2,
        covariance_type='full',
        max_iter=120,
        random_state=42
    )
    gmm.fit(X_c)
    gmm_map_dict[c] = gmm
```

```python
scores = np.zeros((len(X_test_scaled), num_classes))

for i, c in enumerate(range(1, num_classes + 1)):
    ll = gmm_map_dict[c].score_samples(X_test_scaled)         # log P(x | class)
    prior = np.log(class_priors[c])                           # log P(class)
    scores[:, i] = ll + prior                                 # MAP = likelihood + prior

y_pred = np.argmax(scores, axis=1) + 1
```

```python
acc = accuracy_score(y_test, y_pred)
f1  = f1_score(y_test, y_pred, average='weighted')

results_bayes.append({
    "Model": "MAP_GMM_Classifier",
    "Accuracy": acc,
    "Weighted_F1": f1,
    "Time_Sec": None
})

results.append({
    "Model": "MAP_GMM_Classifier",
    "Accuracy": acc,
    "Weighted_F1": f1,
    "Time_Sec": None
})

print("Phase 3E Completed")
```

```python
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)
```

```python
def build_mlp_model(input_dim, hidden_units=[128, 64], dropout_rate=0.2):
    inputs = layers.Input(shape=(input_dim,))
    x = inputs
    for units in hidden_units:
        x = layers.Dense(units, activation='relu')(x)
        x = layers.Dropout(dropout_rate)(x)
    outputs = layers.Dense(7, activation='softmax')(x)
    model = models.Model(inputs, outputs)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

mlp_model = build_mlp_model(input_dim=X_train_mlp.shape[1])

history_mlp = mlp_model.fit(
    X_train_mlp, y_train_cat,
    validation_data=(X_val_mlp, y_val_cat),
    epochs=15,
    batch_size=512,
    callbacks=[early_stop],
    verbose=1
)

# Evaluate with unified function
res_mlp = evaluate_model(
    mlp_model,
    "MLP",
    X_test_mlp,
    y_test,
    is_nn_proba=True
)
results_dl.append(res_mlp)
results.append(res_mlp)

res_mlp
```

```python
def build_cnn_model(input_shape):
    inputs = layers.Input(shape=input_shape)  # (features, 1)
    x = layers.Conv1D(64, kernel_size=3, padding='same', activation='relu')(inputs)
    x = layers.Conv1D(64, kernel_size=3, padding='same', activation='relu')(x)
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(64, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(7, activation='softmax')(x)
    model = models.Model(inputs, outputs)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

cnn_model = build_cnn_model(input_shape=X_train_seq.shape[1:])

history_cnn = cnn_model.fit(
    X_train_seq, y_train_cat,
    validation_data=(X_val_seq, y_val_cat),
    epochs=15,
    batch_size=512,
    callbacks=[early_stop],
    verbose=1
)

res_cnn = evaluate_model(
    cnn_model,
    "CNN1D",
    X_test_seq,
    y_test,
    is_nn_proba=True
)
results_dl.append(res_cnn)
results.append(res_cnn)

res_cnn
```

```python
def build_transformer_model(input_shape, d_model=64, num_heads=2, ff_dim=128, dropout_rate=0.2):
    inputs = layers.Input(shape=input_shape)  # (features, 1)

    # Project to d_model
    x = layers.Dense(d_model)(inputs)

    # Self-attention block
    attn_output = layers.MultiHeadAttention(
        num_heads=num_heads,
        key_dim=d_model // num_heads
    )(x, x)
    x = layers.Add()([x, attn_output])
    x = layers.LayerNormalization(epsilon=1e-6)(x)

    # Feed-forward block
    ff = layers.Dense(ff_dim, activation='relu')(x)
    ff = layers.Dense(d_model)(ff)
    x = layers.Add()([x, ff])
    x = layers.LayerNormalization(epsilon=1e-6)(x)

    # Pool over feature dimension
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dropout(dropout_rate)(x)
    x = layers.Dense(64, activation='relu')(x)
    x = layers.Dropout(dropout_rate)(x)

    outputs = layers.Dense(7, activation='softmax')(x)
    model = models.Model(inputs, outputs)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

transformer_model = build_transformer_model(input_shape=X_train_seq.shape[1:])

history_trans = transformer_model.fit(
    X_train_seq, y_train_cat,
    validation_data=(X_val_seq, y_val_cat),
    epochs=15,
    batch_size=512,
    callbacks=[early_stop],
    verbose=1
)

res_trans = evaluate_model(
    transformer_model,
    "Transformer",
    X_test_seq,
    y_test,
    is_nn_proba=True
)
results_dl.append(res_trans)
results.append(res_trans)

res_trans
```

 Why Random Forest / Boosting outperform Deep Learning on Forest Cover Type Dataset
Although neural networks are powerful function approximators, their performance strongly depends on high-dimensional continuous inputs and very large labeled datasets where hierarchical representations are useful (images, audio, text).
The Forest Cover Type dataset is purely tabular, low-dimensional, and its predictive signal is largely based on threshold interactions between features (e.g., if Elevation > 3000 AND Soil Type = 29 AND Horizontal_Distance_To_Roadways < 1000  Class 2).
Tree-based models, especially Random Forest and Gradient Boosting, are architecturally aligned with this structure:
Advantage	Explanation
Natural threshold decisions	Trees split feature space into rectangular decision regions, matching the structure of the dataset
Handles categorical + binary inputs natively	Soil_Type one-hot encoding provides strong splits without needing embeddings
Robust to scaling	No need for normalization; tree models rely on feature ordering rather than magnitude
Captures feature interactions automatically	Boosted trees find high-order feature combinations without feature engineering
Resistant to noisy / irrelevant features	Random Forest averages many trees, reducing variance
Deep learning models (MLP / CNN / Transformer) struggle because:
Limitation	Reason
Requires many samples per input dimension	Tabular data lacks spatial/temporal patterns
Sensitive to feature scaling & initialization	Training variance is higher than trees
Needs long hyperparameter tuning	Without it, ML models converge sub-optimally
Does not exploit categorical structure	One-hot soil types do not convey ordinal meaning
Therefore, even though neural networks can learn non-linear decision boundaries, the data manifold here benefits more from partitioning rules than hierarchical feature extraction.
 Result: Random Forest / Gradient Boosting consistently outperform Deep Learning on this dataset, both in Accuracy and Weighted F1, while requiring less tuning and training time.
